```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
from io import StringIO

# === ARIMA Explanation ===
# ARIMA (AutoRegressive Integrated Moving Average) models time-series data to forecast future values.
# - AutoRegressive (AR, p): Uses past values of the series (e.g., PDI at t-1, t-2) to predict current values.
# - Integrated (I, d): Applies differencing to remove trends and achieve stationarity (constant mean/variance).
# - Moving Average (MA, q): Uses past forecast errors to model short-term fluctuations.
# We use ARIMA to model log(PDI) because economic series like PDI often have trends and temporal dependencies.
# Log transformation stabilizes variance, making the series more suitable for modeling.

# Step 1: Data Preparation
# Load the dataset provided as a CSV string (quarterly data from 1970.1 to 1991.4).
DATA_STRING = """YEAR,GDP,PDI,PCE,PROFITS,DIVIDENDS
1970.1,2872.8,1990.6,1800.5,44.7,24.5
1970.2,2860.3,2020.1,1807.5,44.4,23.9
1970.3,2896.6,2045.3,1824.7,44.9,23.3
1970.4,2873.7,2045.2,1821.2,42.1,23.1
1971.1,2942.9,2073.9,1849.9,48.8,23.8
1971.2,2947.4,2098,1863.5,50.7,23.7
1971.3,2966,2106.6,1876.9,54.2,23.8
1971.4,2980.8,2121.1,1904.6,55.7,23.7
1972.1,3037.3,2129.7,1929.3,59.4,25
1972.2,3089.7,2149.1,1963.3,60.1,25.5
1972.3,3125.8,2193.9,1989.1,62.8,26.1
1972.4,3175.5,2272,2032.1,68.3,26.5
1973.1,3253.3,2300.7,2063.9,79.1,27
1973.2,3267.6,2315.2,2062,81.2,27.8
1973.3,3264.3,2337.9,2073.7,81.3,28.3
1973.4,3289.1,2382.7,2067.4,85,29.4
1974.1,3259.4,2334.7,2050.8,89,29.8
1974.2,3267.6,2304.5,2059,91.2,30.4
1974.3,3239.1,2315,2065.5,97.1,30.9
1974.4,3226.4,2313.7,2039.9,86.8,30.5
1975.1,3154,2282.5,2051.8,75.8,30
1975.2,3190.4,2390.3,2086.9,81,29.7
1975.3,3249.9,2354.4,2114.4,97.8,30.1
1975.4,3292.5,2389.4,2137,103.4,30.6
1976.1,3356.7,2424.5,2179.3,108.4,32.6
1976.2,3369.2,2434.9,2194.7,109.2,35
1976.3,3381,2444.7,2213,110,36.6
1976.4,3416.3,2459.5,2242,110.3,38.3
1977.1,3466.4,2463,2271.3,121.5,39.2
1977.2,3525,2490.3,2280.8,129.7,40
1977.3,3574.4,2541,2302.6,135.1,41.4
1977.4,3567.2,2556.2,2331.6,134.8,42.4
1978.1,3591.8,2587.3,2347.1,137.5,43.5
1978.2,3707,2631.9,2394,154,44.5
1978.3,3735.6,2653.2,2404.5,158,46.6
1978.4,3779.6,2680.9,2421.6,167.8,48.9
1979.1,3780.8,2699.2,2437.9,168.2,50.5
1979.2,3784.3,2697.6,2435.4,174.1,51.8
1979.3,3807.5,2715.3,2454.7,178.1,52.7
1979.4,3814.6,2728.1,2465.4,173.4,54.5
1980.1,3830.8,2742.9,2464.6,174.3,57.6
1980.2,3732.6,2692,2414.2,144.5,58.7
1980.3,3733.5,2722.5,2440.3,151,59.3
1980.4,3808.5,2777,2469.2,154.6,60.5
1981.1,3860.5,2783.7,2475.5,159.5,64
1981.2,3844.4,2776.7,2476.1,143.7,68.4
1981.3,3864.5,2814.1,2487.4,147.6,71.9
1981.4,3803.1,2808.8,2468.6,140.3,72.4
1982.1,3756.1,2795,2484,114.4,70
1982.2,3771.1,2824.8,2488.9,114,68.4
1982.3,3754.4,2829,2502.5,114.6,69.2
1982.4,3759.6,2832.6,2539.3,109.9,72.5
1983.1,3783.5,2843.6,2556.5,113.6,77
1983.2,3886.5,2867,2604,133,80.5
1983.3,3944.4,2903,2639,145.7,83.1
1984.1,4089.5,3033.2,2703.8,155.1,83.3
1984.2,4144,3065.9,2741.1,152.6,82.2
1984.3,4166.4,3102.7,2754.6,141.8,81.7
1984.4,4194.2,3118.5,2784.8,136.3,83.4
1985.1,4221.8,3123.6,2824.9,125.2,87.2
1985.2,4254.8,3189.6,2849.7,124.8,90.8
1985.3,4309,3156.5,2893.3,129.8,94.1
1985.4,4333.5,3178.7,2895.3,134.2,97.4
1986.1,4390.5,3227.5,2922.4,109.2,105.1
1986.2,4387.7,3281.4,2947.9,106,110.7
1986.3,4412.6,3272.6,2993.7,111,112.3
1986.4,4427.1,3266.2,3012.5,119.2,111
1987.1,4460,3295.2,3011.5,140.2,108
1987.2,4515.3,3241.7,3046.8,157.9,105.5
1987.3,4559.3,3285.7,3075.8,169.1,105.1
1987.4,4625.5,3335.8,3074.6,176,106.3
1988.1,4655.3,3380.1,3128.2,195.5,109.6
1988.2,4704.8,3386.3,3147.8,207.2,113.3
1988.3,4734.5,3407.5,3170.6,213.4,117.5
1988.4,4779.7,3443.1,3202.9,226,121
1989.1,4809.8,3473.9,3200.9,221.3,124.6
1989.2,4832.4,3450.9,3208.6,206.2,127.1
1989.3,4845.6,3466.9,3241.1,195.7,129.1
1989.4,4859.7,3493,3241.6,203,130.7
1990.1,4880.8,3531.4,3258.8,199.1,132.3
1990.2,4900.3,3545.3,3258.6,193.7,132.5
1990.3,4903.3,3547,3281.2,196.3,133.8
1990.4,4855.1,3529.5,3251.8,199,136.2
1991.1,4824,3514.8,3241.1,189.7,137.8
1991.2,4840.7,3537.4,3252.4,182.7,136.7
1991.3,4862.7,3539.9,3271.2,189.6,138.1
1991.4,4868,3547.5,3271.1,190.3,138.5
"""

# Parse data into a DataFrame
data = pd.read_csv(StringIO(DATA_STRING))

# Convert YEAR (e.g., 1970.1) to quarterly datetime (e.g., 1970-01-01 for Q1)
# We map 0.1 to January, 0.2 to April, etc., for quarterly periods
data['date'] = pd.to_datetime(data['YEAR'].apply(lambda x: f"{int(x)}-{int((x % 1) * 12 + 1):02d}-01"))
data.set_index('date', inplace=True)
data.drop('YEAR', axis=1, inplace=True)

# Compute natural logarithm of PDI to stabilize variance
data['log_PDI'] = np.log(data['PDI'])

# Check for missing values and drop them (none expected in this dataset)
data = data.dropna()

# Visualize log(PDI) to inspect trends and potential seasonality
plt.figure(figsize=(10, 6))
plt.plot(data['log_PDI'], label='Log(PDI)')
plt.title('Log of Personal Disposable Income (1987 $ Billions)')
plt.xlabel('Date')
plt.ylabel('Log(PDI)')
plt.legend()
plt.show()

# Step 2: Stationarity Assessment
# Stationarity is critical for ARIMA: a stationary series has constant mean, variance, and no trend.
# We use the Augmented Dickey-Fuller (ADF) test to check stationarity.
# - Null hypothesis: Series has a unit root (non-stationary).
# - If p-value ≤ 0.05, reject null and conclude stationarity.
print('\n=== Stationarity Check for Log(PDI) ===')
adf_result = adfuller(data['log_PDI'])
print('ADF Test on Log(PDI):')
print(f'ADF Statistic: {adf_result[0]:.4f}')
print(f'p-value: {adf_result[1]:.4f}')
print('Critical Values:', adf_result[4])

# Initialize differencing order and series for modeling
d = 0
series_nodiff = data['log_PDI']
is_stationary = adf_result[1] <= 0.05

# If non-stationary, apply first differencing
if not is_stationary:
    print('Log(PDI) is non-stationary (p > 0.05), applying first differencing.')
    data['log_PDI_diff'] = data['log_PDI'].diff()
    series_diff = data['log_PDI_diff'].dropna()
    d = 1
    adf_result_diff = adfuller(series_diff)
    print('ADF Test on Differenced Log(PDI):')
    print(f'ADF Statistic: {adf_result_diff[0]:.4f}')
    print(f'p-value: {adf_result_diff[1]:.4f}')
    print('Critical Values:', adf_result_diff[4])

    # Plot differenced series to visualize stationarity
    plt.figure(figsize=(10, 6))
    plt.plot(series_diff, label='Differenced Log(PDI)')
    plt.title('Differenced Log of Personal Disposable Income')
    plt.xlabel('Date')
    plt.ylabel('Differenced Log(PDI)')
    plt.legend()
    plt.show()
else:
    print('Log(PDI) is stationary (p ≤ 0.05), no differencing needed.')
    series_diff = None

# Function to fit ARIMA model, perform diagnostics, and forecast
def fit_arima(series, d, series_name, original_series=data['log_PDI']):
    """
    Fits ARIMA models, evaluates them, and generates forecasts.
    Parameters:
    - series: Time series to model (non-differenced or differenced).
    - d: Differencing order (0 for non-differenced, 1 for differenced).
    - series_name: Name for labeling outputs (e.g., 'Non-Differenced Log(PDI)').
    - original_series: Original log(PDI) series for consistent model fitting.
    Returns: model, order, AIC, and forecast object.
    """
    # Step 3: Model Identification
    # ACF and PACF plots help identify AR(p) and MA(q) orders.
    # - ACF: Shows correlation of series with its lags; significant spikes suggest MA(q).
    # - PACF: Shows partial correlation after removing earlier lags; spikes suggest AR(p).
    print(f'\n=== Model Identification for {series_name} ===')
    plt.figure(figsize=(12, 6))
    plt.subplot(121)
    plot_acf(series.dropna(), lags=20, ax=plt.gca())
    plt.title(f'ACF of {series_name}')
    plt.subplot(122)
    plot_pacf(series.dropna(), lags=20, ax=plt.gca())
    plt.title(f'PACF of {series_name}')
    plt.tight_layout()
    plt.show()

    # Define candidate ARIMA models to test various (p,d,q) combinations
    # We limit p,q ≤ 2 for simplicity and computational efficiency
    candidate_models = [(0, d, 1), (1, d, 0), (1, d, 1), (2, d, 1), (1, d, 2)]
    best_aic = float('inf')
    best_model = None
    best_order = None

    # Step 4: Model Estimation
    # Fit each candidate model using maximum likelihood estimation
    # Select the model with the lowest AIC (Akaike Information Criterion)
    # AIC balances model fit and complexity (lower is better)
    print(f'\nFitting ARIMA Models for {series_name}')
    for order in candidate_models:
        try:
            # Fit ARIMA to original log(PDI) with specified differencing order
            model = ARIMA(original_series, order=order)
            results = model.fit()
            print(f'ARIMA{order} - AIC: {results.aic:.2f}')
            if results.aic < best_aic:
                best_aic = results.aic
                best_model = results
                best_order = order
        except Exception as e:
            print(f'ARIMA{order} failed to converge: {e}')
            continue

    if best_model is None:
        print(f'No valid ARIMA model found for {series_name}')
        return None, None, None, None

    # Step 5: Model Diagnostics
    # Check residuals (errors) to ensure the model captures all patterns
    # Residuals should be white noise: no autocorrelation, random fluctuations
    print(f'\nDiagnostics for ARIMA{best_order} ({series_name})')
    residuals = best_model.resid
    plt.figure(figsize=(10, 6))
    plt.plot(residuals)
    plt.title(f'Residuals of ARIMA{best_order} for {series_name}')
    plt.xlabel('Date')
    plt.ylabel('Residuals')
    plt.show()

    # Ljung-Box test checks for residual autocorrelation
    # p-values > 0.05 indicate no significant autocorrelation (white noise)
    lb_test = acorr_ljungbox(residuals, lags=[10], return_df=True)
    print(f'Ljung-Box Test for {series_name}:')
    print(lb_test)

    # ACF of residuals should show no significant spikes
    plt.figure(figsize=(10, 6))
    plot_acf(residuals, lags=20)
    plt.title(f'ACF of Residuals for {series_name}')
    plt.show()

    # Step 6: Forecasting
    # Generate 12-quarter (3-year) forecast with confidence intervals
    forecast = best_model.get_forecast(steps=12)
    forecast_mean = forecast.predicted_mean
    forecast_ci = forecast.conf_int()
    forecast_index = pd.date_range(start=data.index[-1] + pd.offsets.QuarterBegin(1), periods=12, freq='QS')

    # Plot forecast alongside observed data
    plt.figure(figsize=(10, 6))
    plt.plot(original_series, label='Observed Log(PDI)')
    plt.plot(forecast_index, forecast_mean, label='Forecast', color='red')
    plt.fill_between(forecast_index, forecast_ci.iloc[:, 0], forecast_ci.iloc[:, 1], color='red', alpha=0.2)
    plt.title(f'Log(PDI) Forecast for {series_name} (ARIMA{best_order})')
    plt.xlabel('Date')
    plt.ylabel('Log(PDI)')
    plt.legend()
    plt.show()

    return best_model, best_order, best_aic, forecast

# Fit ARIMA to non-differenced log(PDI)
print('\n=== Fitting ARIMA to Non-Differenced Log(PDI) ===')
# Model the original log(PDI) series without differencing (d=0)
model_nodiff, order_nodiff, aic_nodiff, forecast_nodiff = fit_arima(series_nodiff, d=0, series_name='Non-Differenced Log(PDI)')

# Fit ARIMA to differenced log(PDI) if non-stationary
if not is_stationary and series_diff is not None:
    print('\n=== Fitting ARIMA to Differenced Log(PDI) ===')
    # Model the differenced series (d=1)
    model_diff, order_diff, aic_diff, forecast_diff = fit_arima(series_diff, d=1, series_name='Differenced Log(PDI)')
else:
    print('\nLog(PDI) is stationary, no differencing needed.')
    model_diff, order_diff, aic_diff, forecast_diff = None, None, None, None

# Step 7: Model Comparison
# Compare non-differenced and differenced models to select the better one
print('\n=== Model Comparison ===')
print('Non-Differenced Model:')
if model_nodiff:
    print(f'Model: ARIMA{order_nodiff}')
    print(f'AIC: {aic_nodiff:.2f}')
    print('Parameters:')
    print(model_nodiff.params)
    lb_pvalue_nodiff = acorr_ljungbox(model_nodiff.resid, lags=[10], return_df=True)['lb_pvalue'].iloc[0]
    print(f'Ljung-Box p-value (lag 10): {lb_pvalue_nodiff:.4f}')

if model_diff:
    print('\nDifferenced Model:')
    print(f'Model: ARIMA{order_diff}')
    print(f'AIC: {aic_diff:.2f}')
    print('Parameters:')
    print(model_diff.params)
    lb_pvalue_diff = acorr_ljungbox(model_diff.resid, lags=[10], return_df=True)['lb_pvalue'].iloc[0]
    print(f'Ljung-Box p-value (lag 10): {lb_pvalue_diff:.4f}')

# Compare forecasts visually if both models exist
if forecast_nodiff is not None and forecast_diff is not None:
    plt.figure(figsize=(10, 6))
    plt.plot(data['log_PDI'], label='Observed Log(PDI)')
    plt.plot(forecast_nodiff.predicted_mean.index, forecast_nodiff.predicted_mean, label='Non-Differenced Forecast', color='red')
    plt.plot(forecast_diff.predicted_mean.index, forecast_diff.predicted_mean, label='Differenced Forecast', color='blue')
    plt.title('Forecast Comparison: Non-Differenced vs Differenced ARIMA')
    plt.xlabel('Date')
    plt.ylabel('Log(PDI)')
    plt.legend()
    plt.show()

# Detailed Comparison Insights
print('\n=== Detailed Comparison Insights ===')
if model_nodiff and model_diff and aic_nodiff and aic_diff:
    print('1. **AIC Comparison**:')
    print(f'   - Non-Differenced AIC: {aic_nodiff:.2f}')
    print(f'   - Differenced AIC: {aic_diff:.2f}')
    if aic_nodiff < aic_diff:
        print('   - Non-differenced model has lower AIC, suggesting better fit and parsimony.')
    else:
        print('   - Differenced model has lower AIC, suggesting better fit after accounting for stationarity.')

    print('\n2. **Residual Diagnostics**:')
    print(f'   - Non-Differenced Ljung-Box p-value: {lb_pvalue_nodiff:.4f}')
    print(f'   - Differenced Ljung-Box p-value: {lb_pvalue_diff:.4f}')
    print('   - p-values > 0.05 indicate residuals are white noise (no autocorrelation).')
    if lb_pvalue_diff > lb_pvalue_nodiff:
        print('   - Differenced model residuals are closer to white noise, suggesting better pattern capture.')
    else:
        print('   - Non-differenced model residuals are closer to white noise.')

    print('\n3. **Stationarity**:')
    print('   - Differenced model ensures stationarity (confirmed by ADF test), which is critical for reliable ARIMA forecasts.')
    print('   - Non-differenced model may fit well but risks instability if the series is non-stationary.')

    print('\n4. **Forecast Plausibility**:')
    print('   - Check the forecast plot: Differenced model forecasts may be more stable (less trend drift).')
    print('   - Confidence intervals indicate forecast uncertainty; narrower intervals suggest more precision.')

    print('\nRecommendation:')
    if aic_diff < aic_nodiff and lb_pvalue_diff > 0.05:
        print('   - Prefer the differenced model: lower AIC and stationary series ensure reliable forecasts.')
    elif aic_nodiff < aic_diff and lb_pvalue_nodiff > 0.05:
        print('   - Prefer the non-differenced model: lower AIC and adequate residuals, but verify stationarity.')
    else:
        print('   - Compare forecast plots and economic context to choose. Differenced model is safer for non-stationary data.')
else:
    print('Only non-differenced model fitted (series is stationary or differenced model failed).')
    print('Use non-differenced model if residuals are white noise (Ljung-Box p > 0.05).')

# Step 8: Decision-Making Insights
print('\n=== Decision-Making Insights ===')
print('ARIMA models predict future PDI, informing economic decisions (e.g., consumer spending trends).')
if model_diff and aic_diff < aic_nodiff and lb_pvalue_diff > 0.05:
    print(f'Best Model: ARIMA{order_diff} (Differenced)')
    print('Use this model for forecasting due to stationarity and better fit.')
elif model_nodiff:
    print(f'Best Model: ARIMA{order_nodiff} (Non-Differenced)')
    print('Use this model if residuals are adequate, but confirm stationarity for long-term forecasts.')
print('Forecasts are in log scale; use np.exp(forecast) to convert to original PDI scale.')
print('Applications:')
print('- Economic Planning: Predict consumer spending capacity based on PDI trends.')
print('- Policy Decisions: Assess impacts of fiscal policies on disposable income.')
print('- Investment: Gauge consumer-driven sectors (e.g., retail) based on PDI forecasts.')
```
